```python
import pandas as pd
import seaborn as sns
from sklearn.linear_model import LinearRegression

df=pd.read_csv('Credit.csv')
df.head(5)
```

## Income vs Limit

```python
plot = sns.lmplot(x='Income',y='Limit', data=df)
```

```python
rl = sns.residplot('Income','Limit', data=df )
rl.set(xlabel='Income', ylabel='Limit Residuals')
```

```python
lin_reg = LinearRegression()
```

```python
lin_reg.fit(df[['Income']], df[['Limit']])
print("Slope is ",lin_reg.coef_)
print("Intercept is ",lin_reg.intercept_)
```
